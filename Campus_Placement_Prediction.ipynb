**Data Preprocessing**

```python
data.head()
```

```python
data=data.drop(['sl_no','salary'],axis=1)
```

```python
data.head(1)
```

**Encoding the Categorical Coloumn**

```python
data['ssc_b'].unique()
```

```python
data['ssc_b']=data['ssc_b'].map({'Central':1,'Others':0})
```

```python
data.head(2)
```

```python
data['hsc_b'].unique()
```

```python
data['hsc_b']=data['hsc_b'].map({'Central':1,'Others':0})
```

```python
data.head(2)
```

```python
data['hsc_s'].unique()
```

```python
data['hsc_s']=data['hsc_s'].map({'Commerce':1,'Arts':0,'Science':2})
```

```python
data.head(2)
```

```python
data['degree_t'].unique()
```

```python
data['degree_t']=data['degree_t'].map({'Sci&Tech':2,'Comm&Mgmt':1,'Others':0})
```

```python
data.head(2)
```

```python
data['specialisation'].unique()
```

```python
data['specialisation']=data['specialisation'].map({'Mkt&HR':1,'Mkt&Fin':0})
```

```python
data.head()
```

```python
data['workex'].unique()
```

```python
data['workex']=data['workex'].map({'Yes':1,'No':0})
data.head(1)
```

```python
data['status'].unique()
```

```python
data['status']=data['status'].map({'Placed':1,'Not Placed':0})
```

```python
data.head(1)
```

**Store Feature Matrix in X and Response(Target)in Vector y**

```python
data.columns
```

```python
x=data.drop('status',axis=1)
y=data['status']
```

**Splitting the Dataset into the training set and Test set**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

**Importing the models**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

**Model Train**

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)

svm=SVC()
svm.fit(x_train,y_train)

knn=KNeighborsClassifier()
knn.fit(x_train,y_train)

dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)

rf=RandomForestClassifier()
rf.fit(x_train,y_train)

gb=GradientBoostingClassifier()
gb.fit(x_train,y_train)
```

**Prediction on the data**

```python
y_pred1=lr.predict(x_test)
y_pred2=svm.predict(x_test)
y_pred3=knn.predict(x_test)
y_pred4=dt.predict(x_test)
y_pred5=rf.predict(x_test)
y_pred6=gb.predict(x_test)
```

**Evaluating the Algorithm**

```python
from sklearn.metrics import accuracy_score
```

```python
score1=accuracy_score(y_test,y_pred1)
score2=accuracy_score(y_test,y_pred2)
score3=accuracy_score(y_test,y_pred3)
score4=accuracy_score(y_test,y_pred4)
score5=accuracy_score(y_test,y_pred5)
score6=accuracy_score(y_test,y_pred6)
```

```python
print(score1,score2,score3,score4,score5,score6)
```

```python
final_data=pd.DataFrame({'Models':['LR','SVC','KNN','DT','RF','GB'],'ACC':[score1*100,
                                                                           score2*100,
                                                                           score3*100,
                                                                           score4*100,
                                                                           score5*100,score6*100]})
```

```python
import seaborn as sns
sns.barplot(x='Models',y='ACC',data=final_data)
```

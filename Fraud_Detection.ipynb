**Online Payment Fraud Detection**

Here we use supervised learning method and will use Decision tree classifier.



```python
#importing library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/Data for Online Fraud Detection.csv')
data.head()
```

```python
#checking null value
data.isnull().sum()
```

```python
# Code to drop null value
data=data.dropna()
#Optionally,reset the index after dropping rows
data.reset_index(drop=True,inplace=True)
print(data)
```

```python
data.isnull().sum()
```

```python
#value of transaction
data.type.value_counts()
```

```python
type=data['type'].value_counts()
transactions=type.index
quantity=type.values

import plotly.express as px
figure=px.pie(data,values=quantity,names=transactions,hole=0.5,title='Distribution of transaction')
figure.show()
```

```python
#mapping in numerical value
data["type"]=data["type"].map({"CASH_OUT":1,"PAYMENT":2,"CASH_IN":3,"TRANSFER":4,"DEBIT":5})
data["isFraud"]=data["isFraud"].map({0:"NO FRAUD",1:"FRAUD"})
data.head()
```

```python
#importing and training the model
from sklearn.model_selection import train_test_split
x=np.array(data[["type","amount","oldbalanceOrg","newbalanceOrig"]])
y=np.array(data[["isFraud"]])
```

```python
from sklearn.tree import DecisionTreeClassifier
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
print(model.score(x_test,y_test))
```

```python
#prediction
#features=[type,amount,oldbalanceorg,newblanceOrig]
features=np.array([[4,9000.60,9000.60,0.0]])
print(model.predict(features))
```
